```python
import csv
import numpy as np


ILDPdataset = []

with open('Indian Liver Patient Dataset (ILPD).csv', 'rb') as csvfile:
    csv_reader = csv.reader(csvfile, delimiter=' ')
    for row in csv_reader:
        for rr in row:
            rrTemp = rr.split(',')
            ILDPdataset.append(rrTemp)
    
csvfile.close()

age_list = [int(row[0]) for row in ILDPdataset] #Age	Age of the patient 
gender_list = [float(0) if row[1]=='Male' else float(1) for row in ILDPdataset] #Gender	Gender of the patient Male = 0 Female = 1
tb_list = [float(row[2]) for row in ILDPdataset] #TB	Total Bilirubin 
db_list = [float(row[3]) for row in ILDPdataset] #DB	Direct Bilirubin
aap_list = [float(row[4]) for row in ILDPdataset] #Alkphos Alkaline Phosphotase
sgpt_list = [float(row[5]) for row in ILDPdataset] #Sgpt Alamine Aminotransferase 
sgot_list = [float(row[6]) for row in ILDPdataset] #Sgot Aspartate Aminotransferase 
tp_list = [float(row[7]) for row in ILDPdataset] #TP	Total Protiens 
alb_list = [float(row[8]) for row in ILDPdataset] #ALB	Albumin 
ag_list = [float (0) if row[9] == '' else float(row[9]) for row in ILDPdataset] #A/G Ratio	Albumin and Globulin Ratio, change null value to 0
class_list = [float(row[10]) for row in ILDPdataset] #Selector field used to split the data into two sets (labeled by the experts) 


# =============================================================================
# Fungsi menghitung nilai entropi dari dua kelas; liver patient dan non liver patient
# menangani error jika ada nilai 0
# =============================================================================

def HD_(p1,p2):
    if p1 == 0 and p2 == 0:
        p1p2_entropi = 0    
    elif p1 == 0:
        p1p2_entropi = -(p1*np.log2(p1, where = False) + p2*np.log2(p2))
    elif p2 == 0:
        p1p2_entropi = -(p1*np.log2(p1) + p2*np.log2(p2, where = False))
    else:   
        p1p2_entropi = -(p1*np.log2(p1) + p2*np.log2(p2))
        
    return p1p2_entropi
     
# =============================================================================
# Konversi atribut Age dengan ketentuan sebagai berikut:
#     Anak-Anak : usia ≤ 11 tahun (a1)
#     Remaja: usia 12 - 20 tahun (a2)
#     Dewasa: usia 21 - 59 tahun (a3)
#     Lansia: ≥ 60 tahun (a4)
# =============================================================================

age_category = []
for xx in age_list:
    if xx <= 11:
        xx = 'a1'
    elif xx <=20:
        xx = 'a2'
    elif xx <= 59:
        xx = 'a3'
    else:
        xx = 'a4'
    age_category.append(xx)
 
    
#print len(age_category)    
# =============================================================================
# Konversi atribut Direct Bilirubin dengan ketentuan:
#     Normal: DB  ≤ 10, selebihnya Tidak Normal
#     Normal (b1), Tidak Normal (b2)
# =============================================================================

db_category = ['b1' if yy <= 10 else 'b2' for yy in db_list]
#print len(db_category)
    
# =============================================================================
# Konversi atribut Total Proteins dengan ketentuan sebagai berikut:
#   Rendah: TP ≤ 3.2 (c1)
#   Normal: 3.3 ≤ TP ≤ 6.4 (c2)
#   Tinggi: TP ≥ 6.5 (c3)
# =============================================================================

tp_category = ['c1' if zz <= 3.2 else 'c3' if zz >= 6.5 else 'c2' for zz in tp_list]

#print len(tp_category)

# =============================================================================
# Konversi atribut Albumin dengan ketentuan sebagai berikut: 
#   Rendah: ALB ≤ 3.2 (d1)
#   Normal: 3.3 ≤ ALB ≤ 6.4 (d2)
#   Tinggi: ALB ≥ 6.5 (d3)
# =============================================================================

alb_category = ['d1' if zz <= 3.2 else 'd3' if zz >= 6.5 else 'd2' for zz in alb_list]

#print len(alb_category)

# =============================================================================
# Array Data 2 hasil klasifikasi
# =============================================================================

print '\n# ================================='
print "# ------------ DATA 2 ------------"
print '# ================================='
data2_list = np.stack((age_category, db_category, tp_category, alb_category, class_list), axis = -1)
print (data2_list)

# =============================================================================
# Pembagian data training dan testing adalah 80:20
# dengan menggunakan teknik random sampling
# =============================================================================

np.random.seed(5)
rand = np.random.rand(len(data2_list)) < 0.8
train_data_list = data2_list[rand]
test_data_list = data2_list[~rand]

print '\nJumlah data testing', test_data_list.shape
print 'Jumlah data training', (train_data_list.shape)

# =============================================================================
# Menghitung nilai entropi dari dua kelas; liver patient dan non liver patient
# =============================================================================
print '\n# ================================='
print "# Mencari H(D) keseluruhan"
print '# ================================='

#print type(train_data_list[0][10])

liver_list = [ii for ii in train_data_list if ii[4] == '1.0']
print 'Jumlah pasien liver = ', len(liver_list)
liver_prob = float(len(liver_list))/float(len(train_data_list))
print 'P(liver) =', liver_prob

nonliver_list = [ii for ii in train_data_list if ii[4] == '2.0']
print '\nJumlah pasien non liver = ', len(nonliver_list)
nonliver_prob = float(len(nonliver_list))/float(len(train_data_list))
print 'P(nonliver) =', nonliver_prob

#HD_entropi = -(liver_prob*np.log2(liver_prob) + nonliver_prob*np.log2(nonliver_prob))
HD_entropi = HD_(liver_prob, nonliver_prob)
print '\nH(D) =', HD_entropi


print '\n# ================================='
print "# Atribut AGE"
print '# ================================='

# =============================================================================
# Mencari jumlah data hasil konversi atribut untuk setiap class (liver (1), non liver(2))
# =============================================================================
a1_liver = np.array([ak for ak in train_data_list if ak[0] == 'a1' and ak[4] == '1.0'])
print '\nJumlah kategori ANAK-ANAK yang LIVER =',len(a1_liver)
a2_liver = np.array([ak for ak in train_data_list if ak[0] == 'a2' and ak[4] == '1.0'])
print 'Jumlah kategori REMAJA yang LIVER =',len(a2_liver)
a3_liver = np.array([ak for ak in train_data_list if ak[0] == 'a3' and ak[4] == '1.0'])
print 'Jumlah kategori DEWASA yang LIVER =',len(a3_liver)
a4_liver = np.array([ak for ak in train_data_list if ak[0] == 'a4' and ak[4] == '1.0'])
print 'Jumlah kategori LANSIA yang LIVER =',len(a4_liver)

a1_nonliver = np.array([ak for ak in train_data_list if ak[0] == 'a1' and ak[4] == '2.0'])
print '\nJumlah kategori ANAK-ANAK yang NONLIVER =',len(a1_nonliver)
a2_nonliver = np.array([ak for ak in train_data_list if ak[0] == 'a2' and ak[4] == '2.0'])
print 'Jumlah kategori REMAJA yang NONLIVER =',len(a2_nonliver)
a3_nonliver = np.array([ak for ak in train_data_list if ak[0] == 'a3' and ak[4] == '2.0'])
print 'Jumlah kategori DEWASA yang NONLIVER =',len(a3_nonliver)
a4_nonliver = np.array([ak for ak in train_data_list if ak[0] == 'a4' and ak[4] == '2.0'])
print 'Jumlah kategori LANSIA yang NONLIVER =',len(a4_nonliver)

# =============================================================================
# Mencari Probabilitas setiap kategori atribut AGE di tiap kelas
# =============================================================================
p_a1_liver = float(len(a1_liver))/(float(len(a1_liver))+float(len(a2_liver))+float(len(a3_liver))+float(len(a4_liver)))
print '\nP(liver|Da1) = ',(p_a1_liver)
p_a2_liver = float(len(a2_liver))/(float(len(a1_liver))+float(len(a2_liver))+float(len(a3_liver))+float(len(a4_liver)))
print 'P(liver|Da2) = ',(p_a2_liver)
p_a3_liver = float(len(a3_liver))/(float(len(a1_liver))+float(len(a2_liver))+float(len(a3_liver))+float(len(a4_liver)))
print 'P(liver|Da3) = ',(p_a3_liver)
p_a4_liver = float(len(a4_liver))/(float(len(a1_liver))+float(len(a2_liver))+float(len(a3_liver))+float(len(a4_liver)))
print 'P(liver|Da4) = ',(p_a4_liver)

p_a1_nonliver = float(len(a1_nonliver))/(float(len(a1_nonliver))+float(len(a2_nonliver))+float(len(a3_nonliver))+float(len(a4_nonliver)))
print '\nP(nonliver|Da1) = ',(p_a1_nonliver)
p_a2_nonliver = float(len(a2_nonliver))/(float(len(a1_nonliver))+float(len(a2_nonliver))+float(len(a3_nonliver))+float(len(a4_nonliver)))
print 'P(nonliver|Da2) = ',(p_a2_nonliver)
p_a3_nonliver = float(len(a3_nonliver))/(float(len(a1_nonliver))+float(len(a2_nonliver))+float(len(a3_nonliver))+float(len(a4_nonliver)))
print 'P(nonliver|Da3) = ',(p_a3_nonliver)
p_a4_nonliver = float(len(a4_nonliver))/(float(len(a1_nonliver))+float(len(a2_nonliver))+float(len(a3_nonliver))+float(len(a4_nonliver)))
print 'P(nonliver|Da4) = ',(p_a4_nonliver)

# =============================================================================
# Menghitung probabilitas tiap kategori (kelas diabaikan)   
# =============================================================================

p_a1 = (float(len(a1_liver))+float(len(a1_nonliver)))/(float(len(train_data_list)))
print '\np(a1) = ', p_a1
p_a2 = (float(len(a2_liver))+float(len(a2_nonliver)))/(float(len(train_data_list)))
print 'p(a2) = ', p_a2
p_a3 = (float(len(a3_liver))+float(len(a3_nonliver)))/(float(len(train_data_list)))
print 'p(a3) = ', p_a3
p_a4 = (float(len(a4_liver))+float(len(a4_nonliver)))/(float(len(train_data_list)))
print 'p(a4) = ', p_a4



# =============================================================================
# Menghitung entropi setiap kategori
# =============================================================================
a1_entropi = HD_(p_a1_liver, p_a1_nonliver)
print '\nH(Da1) = ', a1_entropi
a2_entropi = HD_(p_a2_liver, p_a2_nonliver)
print 'H(Da2) = ', a2_entropi
a3_entropi = HD_(p_a3_liver, p_a3_nonliver)
print 'H(Da3) = ', a3_entropi
a4_entropi = HD_(p_a4_liver, p_a4_nonliver)
print 'H(Da4) = ', a4_entropi

# =============================================================================
# Menghitung nilai entropi dan gain dari atribut AGE
# =============================================================================

age_entropi = a1_entropi*p_a1 + a2_entropi*p_a2 + a3_entropi*p_a3 + a4_entropi*p_a4
print '\nEntropi AGE =', age_entropi
age_ig = HD_entropi - age_entropi
print 'Information Gain AGE =', age_ig


print '\n# ================================='
print "# Atribut Direct Bilirubin (DB)"
print '# ================================='

# =============================================================================
# Mencari jumlah data hasil konversi atribut untuk setiap class (liver (1), non liver(2))
# =============================================================================
b1_liver = np.array([dbk for dbk in train_data_list if dbk[1] == 'b1' and dbk[4] == '1.0'])
print '\nJumlah kategori Direct Bilirubin NORMAL yang LIVER =',len(b1_liver)
b2_liver = np.array([dbk for dbk in train_data_list if dbk[1] == 'b2' and dbk[4] == '1.0'])
print 'Jumlah kategori Direct Bilirubin TIDAK NORMAL yang LIVER =',len(b2_liver)

b1_nonliver = np.array([dbk for dbk in train_data_list if dbk[1] == 'b1' and dbk[4] == '2.0'])
print '\nJumlah kategori Direct Bilirubin NORMAL yang NONLIVER =',len(b1_nonliver)
b2_nonliver = np.array([dbk for dbk in train_data_list if dbk[1] == 'b2' and dbk[4] == '2.0'])
print 'Jumlah kategori Direct Bilirubin TIDAK NORMAL yang NONLIVER =',len(b2_nonliver)

# =============================================================================
# Mencari Probabilitas setiap kategori atribut DB di tiap kelas
# =============================================================================
p_b1_liver = float(len(b1_liver))/(float(len(b1_liver))+float(len(b2_liver)))
print '\nP(liver|Db1) = ',(p_b1_liver)
p_b2_liver = float(len(b2_liver))/(float(len(b1_liver))+float(len(b2_liver)))
print 'P(liver|Db2) = ',(p_b2_liver)

p_b1_nonliver = float(len(b1_nonliver))/(float(len(b1_nonliver))+float(len(b2_nonliver)))
print '\nP(nonliver|Db1) = ',(p_b1_nonliver)
p_b2_nonliver = float(len(b2_nonliver))/(float(len(b1_nonliver))+float(len(b2_nonliver)))
print 'P(nonliver|Db2) = ',(p_b2_nonliver)

# =============================================================================
# Menghitung probabilitas tiap kategori (kelas diabaikan)   
# =============================================================================

p_b1 = (float(len(b1_liver))+float(len(b1_nonliver)))/(float(len(train_data_list)))
print '\np(b1) = ', p_b1
p_b2 = (float(len(b2_liver))+float(len(b2_nonliver)))/(float(len(train_data_list)))
print 'p(b2) = ', p_b2

# =============================================================================
# Menghitung entropi setiap atribut
# =============================================================================
b1_entropi = HD_(p_b1_liver, p_b1_nonliver)
print '\nH(Db1) = ', b1_entropi
#b2_entropi = -(p_b2_liver*np.log2(p_b2_liver) + p_b2_nonliver*np.log2(p_b2_nonliver,where = False))
b2_entropi = HD_(p_b2_liver, p_b2_nonliver)
print 'H(Db2) = ', b2_entropi

# =============================================================================
# Menghitung nilai entropi dan gain dari atribut DB
# =============================================================================

db_entropi = b1_entropi*p_b1 + b2_entropi*p_b2
print '\nEntropi Direct Bilirubin (DB) =', db_entropi
db_ig = HD_entropi - db_entropi
print 'Information Gain Direct Bilirubin (DB) =', db_ig

print '\n# ================================='
print "# Atribut Total Proteins (TP)"
print '# ================================='
# =============================================================================
# Mencari jumlah data hasil konversi atribut untuk setiap class (liver (1), non liver(2))
# =============================================================================

c1_liver = np.array([tpk for tpk in train_data_list if tpk[2] == 'c1' and tpk[4] == '1.0'])
print '\nJumlah kategori Total Proteins RENDAH yang LIVER =',len(c1_liver)
c2_liver = np.array([tpk for tpk in train_data_list if tpk[2] == 'c2' and tpk[4] == '1.0'])
print 'Jumlah kategori Total Proteins NORMAL yang LIVER =',len(c2_liver)
c3_liver = np.array([tpk for tpk in train_data_list if tpk[2] == 'c3' and tpk[4] == '1.0'])
print 'Jumlah kategori Total Proteins TINGGI yang LIVER =',len(c3_liver)

c1_nonliver = np.array([tpk for tpk in train_data_list if tpk[2] == 'c1' and tpk[4] == '2.0'])
print '\nJumlah kategori Total Proteins RENDAH yang NONLIVER =',len(c1_nonliver)
c2_nonliver = np.array([tpk for tpk in train_data_list if tpk[2] == 'c2' and tpk[4] == '2.0'])
print 'Jumlah kategori Total Proteins NORMAL yang NONLIVER =',len(c2_nonliver)
c3_nonliver = np.array([tpk for tpk in train_data_list if tpk[2] == 'c3' and tpk[4] == '2.0'])
print 'Jumlah kategori Total Proteins TINGGI yang NONLIVER =',len(c3_nonliver)

# =============================================================================
# Mencari Probabilitas setiap kategori atribut TP di tiap kelas
# =============================================================================
p_c1_liver = float(len(c1_liver))/(float(len(c1_liver))+float(len(c2_liver))+float(len(c3_liver)))
print '\nP(liver|Dc1) = ',(p_c1_liver)
p_c2_liver = float(len(c2_liver))/(float(len(c1_liver))+float(len(c2_liver))+float(len(c3_liver)))
print 'P(liver|Dc2) = ',(p_c2_liver)
p_c3_liver = float(len(c3_liver))/(float(len(c1_liver))+float(len(c2_liver))+float(len(c3_liver)))
print 'P(liver|Dc3) = ',(p_c3_liver)

p_c1_nonliver = float(len(c1_nonliver))/(float(len(c1_nonliver))+float(len(c2_nonliver))+float(len(c3_nonliver)))
print '\nP(nonliver|Dc1) = ',(p_c1_nonliver)
p_c2_nonliver = float(len(c2_nonliver))/(float(len(c1_nonliver))+float(len(c2_nonliver))+float(len(c3_nonliver)))
print 'P(nonliver|Dc2) = ',(p_c2_nonliver)
p_c3_nonliver = float(len(c3_nonliver))/(float(len(c1_nonliver))+float(len(c2_nonliver))+float(len(c3_nonliver)))
print 'P(nonliver|Dc3) = ',(p_c3_nonliver)

# =============================================================================
# Menghitung probabilitas tiap kategori (kelas diabaikan)   
# =============================================================================

p_c1 = (float(len(c1_liver))+float(len(c1_nonliver)))/(float(len(train_data_list)))
print '\np(c1) = ', p_c1
p_c2 = (float(len(c2_liver))+float(len(c2_nonliver)))/(float(len(train_data_list)))
print 'p(c2) = ', p_c2
p_c3 = (float(len(c3_liver))+float(len(c3_nonliver)))/(float(len(train_data_list)))
print 'p(c3) = ', p_c3

# =============================================================================
# Menghitung entropi setiap atribut
# =============================================================================
c1_entropi = HD_(p_c1_liver, p_c1_nonliver)
print '\nH(Dc1) = ', c1_entropi
c2_entropi = HD_(p_c2_liver, p_c2_nonliver)
print 'H(Dc2) = ', c2_entropi
c3_entropi = HD_(p_c3_liver, p_c3_nonliver)
print 'H(Dc3) = ', c3_entropi

# =============================================================================
# Menghitung nilai entropi dan gain dari atribut TP
# =============================================================================

tp_entropi = c1_entropi*p_c1 + c2_entropi*p_c2 + c3_entropi*p_c3
print '\nEntropi Total Proteins (TP) =', tp_entropi
tp_ig = HD_entropi - tp_entropi
print 'Information Gain Total Proteins (TP) =', tp_ig

print '\n# ================================='
print "# Atribut Albumin (ALB)"
print '# ================================='
# =============================================================================
# Mencari jumlah data hasil konversi atribut untuk setiap class (liver (1), non liver(2))
# =============================================================================
d1_liver = np.array([albk for albk in train_data_list if albk[3] == 'd1' and albk[4] == '1.0'])
print '\nJumlah kategori Albumin RENDAH yang LIVER =',len(d1_liver)
d2_liver = np.array([albk for albk in train_data_list if albk[3] == 'd2' and albk[4] == '1.0'])
print 'Jumlah kategori Albumin NORMAL yang LIVER =',len(d2_liver)
d3_liver = np.array([albk for albk in train_data_list if albk[3] == 'd3' and albk[4] == '1.0'])
print 'Jumlah kategori Albumin TINGGI yang LIVER =',len(d3_liver)

d1_nonliver = np.array([albk for albk in train_data_list if albk[3] == 'd1' and albk[4] == '2.0'])
print '\nJumlah kategori Albumin RENDAH yang NONLIVER =',len(d1_nonliver)
d2_nonliver = np.array([albk for albk in train_data_list if albk[3] == 'd2' and albk[4] == '2.0'])
print 'Jumlah kategori Albumin NORMAL yang NONLIVER =',len(d2_nonliver)
d3_nonliver = np.array([albk for albk in train_data_list if albk[3] == 'd3' and albk[4] == '2.0'])
print 'Jumlah kategori Albumin TINGGI yang NONLIVER =',len(d3_nonliver)

# =============================================================================
# Mencari Probabilitas setiap kategori atribut ALB di tiap kelas
# =============================================================================
p_d1_liver = float(len(d1_liver))/(float(len(d1_liver))+float(len(d2_liver))+float(len(d3_liver)))
print '\nP(liver|Dd1) = ',(p_d1_liver)
p_d2_liver = float(len(d2_liver))/(float(len(d1_liver))+float(len(d2_liver))+float(len(d3_liver)))
print 'P(liver|Dd2) = ',(p_d2_liver)
p_d3_liver = float(len(d3_liver))/(float(len(d1_liver))+float(len(d2_liver))+float(len(d3_liver)))
print 'P(liver|Dd3) = ',(p_d3_liver)

p_d1_nonliver = float(len(d1_nonliver))/(float(len(d1_nonliver))+float(len(d2_nonliver))+float(len(d3_nonliver)))
print '\nP(nonliver|Dd1) = ',(p_d1_nonliver)
p_d2_nonliver = float(len(d2_nonliver))/(float(len(d1_nonliver))+float(len(d2_nonliver))+float(len(d3_nonliver)))
print 'P(nonliver|Dd2) = ',(p_d2_nonliver)
p_d3_nonliver = float(len(d3_nonliver))/(float(len(d1_nonliver))+float(len(d2_nonliver))+float(len(d3_nonliver)))
print 'P(nonliver|Dd3) = ',(p_d3_nonliver)

# =============================================================================
# Menghitung probabilitas tiap kategori (kelas diabaikan)   
# =============================================================================
p_d1 = (float(len(d1_liver))+float(len(d1_nonliver)))/(float(len(train_data_list)))
print '\np(d1) = ', p_d1
p_d2 = (float(len(d2_liver))+float(len(d2_nonliver)))/(float(len(train_data_list)))
print 'p(d2) = ', p_d2
p_d3 = (float(len(d3_liver))+float(len(d3_nonliver)))/(float(len(train_data_list)))
print 'p(d3) = ', p_d3

# =============================================================================
# Menghitung entropi setiap atribut
# =============================================================================
d1_entropi = HD_(p_d1_liver, p_d1_nonliver)
print '\nH(Dd1) = ', d1_entropi
d2_entropi = HD_(p_d2_liver, p_d2_nonliver)
print 'H(Dd2) = ', d2_entropi
d3_entropi = HD_(p_d3_liver, p_d3_nonliver)
print 'H(Dd3) = ', d3_entropi

# =============================================================================
# Menghitung nilai entropi dan gain dari atribut ALB
# =============================================================================

alb_entropi = d1_entropi*p_d1 + d2_entropi*p_d2 + d3_entropi*p_d3
print '\nEntropi Albumin (ALB) =', tp_entropi
alb_ig = HD_entropi - alb_entropi
print 'Information Gain Albumin (ALB) =', alb_ig

print '\n=========================================================================='
print 'Information Gain tertinggi adalah Bilirubin (DB), sehingga DB menjadi root'
print '=========================================================================='

# =============================================================================
# Mencari Split Data berikutnya
# =============================================================================

print '\n# ================================='
print "# Mencari H(D) DB NORMAL"
print '# ================================='

jumlah_b1 = len(b1_liver) + len(b1_nonliver)
print '\nJumlah pasien DB NORMAL = ', jumlah_b1

print '\nJumlah pasien DB NORMAL liver = ', len(b1_liver)
print 'Jumlah pasien DB NORMAL non liver = ', len(b1_nonliver)

b1_liver_prob = float(len(b1_liver))/float(jumlah_b1)
print '\nP(b1_liver) =', b1_liver_prob

b1_nonliver_prob = float(len(b1_nonliver))/float(jumlah_b1)
print 'P(b1_nonliver) =', b1_nonliver_prob


HD_b1_entropi = HD_(b1_liver_prob, b1_nonliver_prob)
print '\nH(b1_D) =', HD_b1_entropi

print '\n# ================================='
print "# Mencari H(D) DB TIDAK NORMAL"
print '# ================================='

jumlah_b2 = len(b2_liver) + len(b2_nonliver)
print '\nJumlah pasien DB TIDAK NORMAL = ', jumlah_b2

print '\nJumlah pasien DB TIDAK NORMAL liver = ', len(b2_liver)
print 'Jumlah pasien DB TIDAK NORMAL non liver = ', len(b2_nonliver)

b2_liver_prob = float(len(b2_liver))/float(jumlah_b2)
print '\nP(b2_liver) =', b2_liver_prob

b2_nonliver_prob = float(len(b2_nonliver))/float(jumlah_b2)
print 'P(b2_nonliver) =', b2_nonliver_prob


HD_b2_entropi = HD_(b2_liver_prob, b2_nonliver_prob)
print '\nH(b2_D) =', HD_b2_entropi

print '\n====================================================================================='
print 'Dari hasil pencarian H(D) dari data DB Normal dan Tidak Normal dapat dilihat bahwa,'
print 'DB Normal harus displit kembali dan DB Tidak Normal majority ke target kelas LIVER'
print '====================================================================================='

# =============================================================================
# Atribut Age
# =============================================================================

print '\n# ================================='
print "# Atribut AGE, DB NORMAL"
print '# ================================='

a1_b1_liver = np.array([dt2 for dt2 in train_data_list if dt2[0] == 'a1' and dt2[1] == 'b1' and dt2[4] == '1.0'])
print '\nJumlah kategori ANAK-ANAK (a1), Direct Bilirubin (DB) NORMAL (b1), LIVER (1) =', len(a1_b1_liver)
a2_b1_liver = np.array([dt2 for dt2 in train_data_list if dt2[0] == 'a2' and dt2[1] == 'b1' and dt2[4] == '1.0'])
print 'Jumlah kategori REMAJA (a2), Direct Bilirubin (DB) NORMAL (b1), LIVER (1) =', len(a2_b1_liver)
a3_b1_liver = np.array([dt2 for dt2 in train_data_list if dt2[0] == 'a3' and dt2[1] == 'b1' and dt2[4] == '1.0'])
print 'Jumlah kategori DEWASA (a3), Direct Bilirubin (DB) NORMAL (b1), LIVER (1) =', len(a3_b1_liver)
a4_b1_liver = np.array([dt2 for dt2 in train_data_list if dt2[0] == 'a4' and dt2[1] == 'b1' and dt2[4] == '1.0'])
print 'Jumlah kategori LANSIA (a4), Direct Bilirubin (DB) NORMAL (b1), LIVER (1) =', len(a4_b1_liver)

a1_b1_nonliver = np.array([dt2 for dt2 in train_data_list if dt2[0] == 'a1' and dt2[1] == 'b1' and dt2[4] == '2.0'])
print '\nJumlah kategori ANAK-ANAK (a1), Direct Bilirubin (DB) NORMAL (b1), NONLIVER (2) =', len(a1_b1_nonliver)
a2_b1_nonliver = np.array([dt2 for dt2 in train_data_list if dt2[0] == 'a2' and dt2[1] == 'b1' and dt2[4] == '2.0'])
print 'Jumlah kategori REMAJA (a2), Direct Bilirubin (DB) NORMAL (b1), NONLIVER (2) =', len(a2_b1_nonliver)
a3_b1_nonliver = np.array([dt2 for dt2 in train_data_list if dt2[0] == 'a3' and dt2[1] == 'b1' and dt2[4] == '2.0'])
print 'Jumlah kategori DEWASA (a3), Direct Bilirubin (DB) NORMAL (b1), NONLIVER (2) =', len(a3_b1_nonliver)
a4_b1_nonliver = np.array([dt2 for dt2 in train_data_list if dt2[0] == 'a4' and dt2[1] == 'b1' and dt2[4] == '2.0'])
print 'Jumlah kategori LANSIA (a4), Direct Bilirubin (DB) NORMAL (b1), NONLIVER (2) =', len(a4_b1_nonliver)

# =============================================================================
# Mencari Probabilitas setiap kategori atribut AGE di tiap kelas
# =============================================================================
p_a1_b1_liver = float(len(a1_b1_liver))/(float(len(a1_b1_liver))+float(len(a2_b1_liver))+float(len(a3_b1_liver))+float(len(a4_b1_liver)))
print '\nP(b1_liver|Da1) = ',(p_a1_b1_liver)
p_a2_b1_liver = float(len(a2_b1_liver))/(float(len(a1_b1_liver))+float(len(a2_b1_liver))+float(len(a3_b1_liver))+float(len(a4_b1_liver)))
print 'P(b1_liver|Da2) = ',(p_a2_b1_liver)
p_a3_b1_liver = float(len(a3_b1_liver))/(float(len(a1_b1_liver))+float(len(a2_b1_liver))+float(len(a3_b1_liver))+float(len(a4_b1_liver)))
print 'P(b1_liver|Da3) = ',(p_a3_b1_liver)
p_a4_b1_liver = float(len(a4_b1_liver))/(float(len(a1_b1_liver))+float(len(a2_b1_liver))+float(len(a3_b1_liver))+float(len(a4_b1_liver)))
print 'P(b1_liver|Da4) = ',(p_a4_b1_liver)


p_a1_b1_nonliver = float(len(a1_b1_nonliver))/(float(len(a1_b1_nonliver))+float(len(a2_b1_nonliver))+float(len(a3_b1_nonliver))+float(len(a4_b1_nonliver)))
print '\nP(b1_nonliver|Da1) = ',(p_a1_b1_nonliver)
p_a2_b1_nonliver = float(len(a2_b1_nonliver))/(float(len(a1_b1_nonliver))+float(len(a2_b1_nonliver))+float(len(a3_b1_nonliver))+float(len(a4_b1_nonliver)))
print 'P(b1_nonliver|Da2) = ',(p_a2_b1_nonliver)
p_a3_b1_nonliver = float(len(a3_b1_nonliver))/(float(len(a1_b1_nonliver))+float(len(a2_b1_nonliver))+float(len(a3_b1_nonliver))+float(len(a4_b1_nonliver)))
print 'P(b1_nonliver|Da3) = ',(p_a3_b1_nonliver)
p_a4_b1_nonliver = float(len(a4_b1_nonliver))/(float(len(a1_b1_nonliver))+float(len(a2_b1_nonliver))+float(len(a3_b1_nonliver))+float(len(a4_b1_nonliver)))
print 'P(b1_nonliver|Da4) = ',(p_a4_b1_nonliver)

# =============================================================================
# Menghitung entropi setiap atribut
# =============================================================================
a1_b1_entropi = HD_(p_a1_b1_liver, p_a1_b1_nonliver)
print '\nH(Da1_b1) = ', a1_b1_entropi
a2_b1_entropi = HD_(p_a2_b1_liver, p_a2_b1_nonliver)
print 'H(Da2_b1) = ', a2_b1_entropi
a3_b1_entropi = HD_(p_a3_b1_liver, p_a3_b1_nonliver)
print 'H(Da3_b1) = ', a3_b1_entropi
a4_b1_entropi = HD_(p_a4_b1_liver, p_a4_b1_nonliver)
print 'H(Da4_b1) = ', a4_b1_entropi

# =============================================================================
# Menghitung probabilitas tiap kategori (kelas diabaikan)   
# =============================================================================

p_a1_b1 = (float(len(a1_b1_liver))+float(len(a1_b1_nonliver)))/(float(jumlah_b1))
print '\np(a1_b1) = ', p_a1_b1
p_a2_b1 = (float(len(a2_b1_liver))+float(len(a2_b1_nonliver)))/(float(jumlah_b1))
print 'p(a2_b1) = ', p_a2_b1
p_a3_b1 = (float(len(a3_b1_liver))+float(len(a3_b1_nonliver)))/(float(jumlah_b1))
print 'p(a3_b1) = ', p_a3_b1
p_a4_b1 = (float(len(a4_b1_liver))+float(len(a4_b1_nonliver)))/(float(jumlah_b1))
print 'p(a4_b1) = ', p_a4_b1

# =============================================================================
# Menghitung nilai entropi dan gain dari atribut AGE
# =============================================================================

age_b1_entropi = a1_b1_entropi*p_a1_b1 + a2_b1_entropi*p_a2_b1 + a3_b1_entropi*p_a3_b1 + a4_b1_entropi*p_a4_b1
print '\nEntropi AGE b1 =', age_b1_entropi
age_b1_ig = HD_b1_entropi - age_b1_entropi
print 'Information Gain AGE b1 =', age_b1_ig


# =============================================================================
# Atribut Total Proteins (TP)
# =============================================================================

print '\n# =============================================='
print "# Atribut Total Proteins (TP), DB NORMAL"
print '# ============================================='

c1_b1_liver = np.array([dt2 for dt2 in train_data_list if dt2[2] == 'c1' and dt2[1] == 'b1' and dt2[4] == '1.0'])
print '\nJumlah kategori Total Proteins RENDAH, Direct Bilirubin (DB) NORMAL (b1), LIVER (1) =',len(c1_b1_liver)
c2_b1_liver = np.array([dt2 for dt2 in train_data_list if dt2[2] == 'c2' and dt2[1] == 'b1' and dt2[4] == '1.0'])
print 'Jumlah kategori Total Proteins NORMAL, Direct Bilirubin (DB) NORMAL (b1), LIVER (1) =',len(c2_b1_liver)
c3_b1_liver = np.array([dt2 for dt2 in train_data_list if dt2[2] == 'c3' and dt2[1] == 'b1' and dt2[4] == '1.0'])
print 'Jumlah kategori Total Proteins TINGGI, Direct Bilirubin (DB) NORMAL (b1), LIVER (1) =',len(c3_b1_liver)

c1_b1_nonliver = np.array([dt2 for dt2 in train_data_list if dt2[2] == 'c1' and dt2[1] == 'b1' and dt2[4] == '2.0'])
print '\nJumlah kategori Total Proteins RENDAH, Direct Bilirubin (DB) NORMAL (b1), NONLIVER (2) =',len(c1_b1_nonliver)
c2_b1_nonliver = np.array([dt2 for dt2 in train_data_list if dt2[2] == 'c2' and dt2[1] == 'b1' and dt2[4] == '2.0'])
print 'Jumlah kategori Total Proteins NORMAL, Direct Bilirubin (DB) NORMAL (b1), NONLIVER (2) =',len(c2_b1_nonliver)
c3_b1_nonliver = np.array([dt2 for dt2 in train_data_list if dt2[2] == 'c3' and dt2[1] == 'b1' and dt2[4] == '2.0'])
print 'Jumlah kategori Total Proteins TINGGI, Direct Bilirubin (DB) NORMAL (b1), NONLIVER (2) =',len(c3_b1_nonliver)

# =============================================================================
# Mencari Probabilitas setiap kategori atribut TP di tiap kelas
# =============================================================================
p_c1_b1_liver = float(len(c1_b1_liver))/(float(len(c1_b1_liver))+float(len(c2_b1_liver))+float(len(c3_b1_liver)))
print '\nP(b1_liver|Dc1) = ',(p_c1_b1_liver)
p_c2_b1_liver = float(len(c2_b1_liver))/(float(len(c1_b1_liver))+float(len(c2_b1_liver))+float(len(c3_b1_liver)))
print 'P(b1_liver|Dc2) = ',(p_c2_b1_liver)
p_c3_b1_liver = float(len(c3_b1_liver))/(float(len(c1_b1_liver))+float(len(c2_b1_liver))+float(len(c3_b1_liver)))
print 'P(b1_liver|Dc3) = ',(p_c3_b1_liver)

p_c1_b1_nonliver = float(len(c1_b1_nonliver))/(float(len(c1_b1_nonliver))+float(len(c2_b1_nonliver))+float(len(c3_b1_nonliver)))
print '\nP(b1_nonliver|Dc1) = ',(p_c1_b1_nonliver)
p_c2_b1_nonliver = float(len(c2_b1_nonliver))/(float(len(c1_b1_nonliver))+float(len(c2_b1_nonliver))+float(len(c3_b1_nonliver)))
print 'P(b1_nonliver|Dc2) = ',(p_c2_b1_nonliver)
p_c3_b1_nonliver = float(len(c3_b1_nonliver))/(float(len(c1_b1_nonliver))+float(len(c2_b1_nonliver))+float(len(c3_b1_nonliver)))
print 'P(b1_nonliver|Dc3) = ',(p_c3_b1_nonliver)

# =============================================================================
# Menghitung entropi setiap atribut
# =============================================================================
c1_b1_entropi = HD_(p_c1_b1_liver, p_c1_b1_nonliver)
print '\nH(Dc1_b1) = ', c1_b1_entropi
c2_b1_entropi = HD_(p_c2_b1_liver, p_c2_b1_nonliver)
print 'H(Dc2_b1) = ', c2_b1_entropi
c3_b1_entropi = HD_(p_c3_b1_liver, p_c3_b1_nonliver)
print 'H(Dc3_b1) = ', c3_b1_entropi

# =============================================================================
# Menghitung probabilitas tiap kategori (kelas diabaikan)   
# =============================================================================

p_c1_b1 = (float(len(c1_b1_liver))+float(len(c1_b1_nonliver)))/(float(jumlah_b1))
print '\np(c1_b1) = ', p_c1_b1
p_c2_b1 = (float(len(c2_b1_liver))+float(len(c2_b1_nonliver)))/(float(jumlah_b1))
print 'p(c2_b1) = ', p_c2_b1
p_c3_b1 = (float(len(c3_b1_liver))+float(len(c3_b1_nonliver)))/(float(jumlah_b1))
print 'p(c3_b1) = ', p_c3_b1

# =============================================================================
# Menghitung nilai entropi dan gain dari atribut TP
# =============================================================================

tp_b1_entropi = c1_b1_entropi*p_c1_b1 + c2_b1_entropi*p_c2_b1 + c3_b1_entropi*p_c3_b1
print '\nEntropi Total Proteins (TP) b1 =', tp_b1_entropi
tp_b1_ig = HD_b1_entropi - tp_b1_entropi
print 'Information Gain Total Proteins (TP) b1 =', tp_b1_ig

# =============================================================================
# Atribut Albumin (ALB)
# =============================================================================

print '\n# ======================================='
print "# Atribut Albumin (ALB), DB NORMAL"
print '# ======================================='

d1_b1_liver = np.array([dt2 for dt2 in train_data_list if dt2[3] == 'd1' and dt2[1] == 'b1' and dt2[4] == '1.0'])
print '\nJumlah kategori Total Proteins RENDAH, Albumin (ALB) NORMAL (b1), LIVER (1) =',len(d1_b1_liver)
d2_b1_liver = np.array([dt2 for dt2 in train_data_list if dt2[3] == 'd2' and dt2[1] == 'b1' and dt2[4] == '1.0'])
print 'Jumlah kategori Total Proteins NORMAL, Albumin (ALB) NORMAL (b1), LIVER (1) =',len(d2_b1_liver)
d3_b1_liver = np.array([dt2 for dt2 in train_data_list if dt2[3] == 'd3' and dt2[1] == 'b1' and dt2[4] == '1.0'])
print 'Jumlah kategori Total Proteins TINGGI, Albumin (ALB) NORMAL (b1), LIVER (1) =',len(d3_b1_liver)

d1_b1_nonliver = np.array([dt2 for dt2 in train_data_list if dt2[3] == 'd1' and dt2[1] == 'b1' and dt2[4] == '2.0'])
print '\nJumlah kategori Total Proteins RENDAH, Albumin (ALB) NORMAL (b1), NONLIVER (2) =',len(d1_b1_nonliver)
d2_b1_nonliver = np.array([dt2 for dt2 in train_data_list if dt2[3] == 'd2' and dt2[1] == 'b1' and dt2[4] == '2.0'])
print 'Jumlah kategori Total Proteins NORMAL, Albumin (ALB) NORMAL (b1), NONLIVER (2) =',len(d2_b1_nonliver)
d3_b1_nonliver = np.array([dt2 for dt2 in train_data_list if dt2[3] == 'd3' and dt2[1] == 'b1' and dt2[4] == '2.0'])
print 'Jumlah kategori Total Proteins TINGGI, Albumin (ALB) NORMAL (b1), NONLIVER (2) =',len(d3_b1_nonliver)

# =============================================================================
# Mencari Probabilitas setiap kategori atribut ALB di tiap kelas
# =============================================================================
p_d1_b1_liver = float(len(d1_b1_liver))/(float(len(d1_b1_liver))+float(len(d2_b1_liver))+float(len(d3_b1_liver)))
print '\nP(b1_liver|Dd1) = ',(p_d1_b1_liver)
p_d2_b1_liver = float(len(d2_b1_liver))/(float(len(d1_b1_liver))+float(len(d2_b1_liver))+float(len(d3_b1_liver)))
print 'P(b1_liver|Dd2) = ',(p_d2_b1_liver)
p_d3_b1_liver = float(len(d3_b1_liver))/(float(len(d1_b1_liver))+float(len(d2_b1_liver))+float(len(d3_b1_liver)))
print 'P(b1_liver|Dd3) = ',(p_d3_b1_liver)

p_d1_b1_nonliver = float(len(d1_b1_nonliver))/(float(len(d1_b1_nonliver))+float(len(d2_b1_nonliver))+float(len(d3_b1_nonliver)))
print '\nP(b1_nonliver|Dd1) = ',(p_d1_b1_nonliver)
p_d2_b1_nonliver = float(len(d2_b1_nonliver))/(float(len(d1_b1_nonliver))+float(len(d2_b1_nonliver))+float(len(d3_b1_nonliver)))
print 'P(b1_nonliver|Dd2) = ',(p_d2_b1_nonliver)
p_d3_b1_nonliver = float(len(d3_b1_nonliver))/(float(len(d1_b1_nonliver))+float(len(d2_b1_nonliver))+float(len(d3_b1_nonliver)))
print 'P(b1_nonliver|Dd3) = ',(p_d3_b1_nonliver)

# =============================================================================
# Menghitung entropi setiap atribut
# =============================================================================
d1_b1_entropi = HD_(p_d1_b1_liver, p_d1_b1_nonliver)
print '\nH(Dd1_b1) = ', d1_b1_entropi
d2_b1_entropi = HD_(p_d2_b1_liver, p_d2_b1_nonliver)
print 'H(Dd2_b1) = ', d2_b1_entropi
d3_b1_entropi = HD_(p_d3_b1_liver, p_d3_b1_nonliver)
print 'H(Dd3_b1) = ', d3_b1_entropi

# =============================================================================
# Menghitung probabilitas tiap kategori (kelas diabaikan)   
# =============================================================================

p_d1_b1 = (float(len(d1_b1_liver))+float(len(d1_b1_nonliver)))/(float(jumlah_b1))
print '\np(d1_b1) = ', p_d1_b1
p_d2_b1 = (float(len(d2_b1_liver))+float(len(d2_b1_nonliver)))/(float(jumlah_b1))
print 'p(d2_b1) = ', p_d2_b1
p_d3_b1 = (float(len(d3_b1_liver))+float(len(d3_b1_nonliver)))/(float(jumlah_b1))
print 'p(d3_b1) = ', p_d3_b1

# =============================================================================
# Menghitung nilai entropi dan gain dari atribut ALB
# =============================================================================

alb_b1_entropi = d1_b1_entropi*p_d1_b1 + d2_b1_entropi*p_d2_b1 + d3_b1_entropi*p_d3_b1
print '\nEntropi Albumin (ALB) b1 =', alb_b1_entropi
alb_b1_ig = HD_b1_entropi - alb_b1_entropi
print 'Information Gain Albumin (ALB) b1 =', alb_b1_ig

print '\nInformation Gain tertinggi untuk Bilirubin (DB) NORMAL adalah AGE, sehingga AGE menjadi node berikutnya'
```

```python
import csv
import numpy as np
from sklearn import tree
from sklearn.preprocessing import LabelEncoder
import graphviz

ILDPdataset = []

with open('Indian Liver Patient Dataset (ILPD).csv', 'rb') as csvfile:
    csv_reader = csv.reader(csvfile, delimiter=' ')
    for row in csv_reader:
        for rr in row:
            rrTemp = rr.split(',')
            ILDPdataset.append(rrTemp)
    
csvfile.close()

age_list = [int(row[0]) for row in ILDPdataset] #Age	Age of the patient 
gender_list = [float(0) if row[1]=='Male' else float(1) for row in ILDPdataset] #Gender	Gender of the patient Male = 0 Female = 1
tb_list = [float(row[2]) for row in ILDPdataset] #TB	Total Bilirubin 
db_list = [float(row[3]) for row in ILDPdataset] #DB	Direct Bilirubin
aap_list = [float(row[4]) for row in ILDPdataset] #Alkphos Alkaline Phosphotase
sgpt_list = [float(row[5]) for row in ILDPdataset] #Sgpt Alamine Aminotransferase 
sgot_list = [float(row[6]) for row in ILDPdataset] #Sgot Aspartate Aminotransferase 
tp_list = [float(row[7]) for row in ILDPdataset] #TP	Total Protiens 
alb_list = [float(row[8]) for row in ILDPdataset] #ALB	Albumin 
ag_list = [float (0) if row[9] == '' else float(row[9]) for row in ILDPdataset] #A/G Ratio	Albumin and Globulin Ratio, change null value to 0
class_list = [float(row[10]) for row in ILDPdataset] #Selector field used to split the data into two sets (labeled by the experts) 


print '\n# ================================='
print "# ------------ DATA 1 ------------"
print '# ================================='
data1_list = np.stack((age_list, db_list, tp_list, alb_list, class_list), axis = -1)
print (data1_list)

np.random.seed(5)
rand = np.random.rand(len(data1_list)) < 0.8
train_data_list = data1_list[rand]
test_data_list = data1_list[~rand]

#print test_data_list.shape
#print (train_data_list.shape)

clf = tree.DecisionTreeClassifier(criterion = 'entropy').fit(train_data_list[:,:4],train_data_list[:,4:5])
print clf

prediction = clf.predict(test_data_list[:,:4])

#print prediction
print("The prediction accuracy is: ",clf.score(train_data_list[:,:4],train_data_list[:,4:5])*100,"%")

dot_data = tree.export_graphviz(clf, out_file=None) 
graph = graphviz.Source(dot_data)
graph
```
