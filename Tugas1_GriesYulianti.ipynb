```python
#1.

from sklearn import datasets

iris = datasets.load_iris()

dataIris = iris.data[:]

print "Data matrik n x d"
print dataIris.shape

print "Banyak data adalah n = 150"

print "Dimensi adalah jumlah atribut (d) = 4"
```

```python
#2.

def mean(listKolom):
    meanList = sum(listKolom)/len(listKolom)
    
    return meanList

def varian(listKolom,meanX):
    list_operasiX = []
    for x in listKolom:
        operasiX = (x-meanX)**2
        list_operasiX.append(operasiX)
    
    varianX = sum(list_operasiX)/len(listKolom)
    return varianX

# --- Sepal Length

sepal_length = [row[0] for row in dataIris] 
#sepal_length = dataIris[:,:1] 
#print (sepal_length)
#print type(sepal_length)

mean_sepal_length = mean(sepal_length)
print "\nMean Sepal Length"
print mean_sepal_length

var_sepal_length = varian(sepal_length,mean_sepal_length)
print "Varian Sepal Length"
print var_sepal_length
    

# --- Sepal Width

sepal_width = [row[1] for row in dataIris] 
#print (sepal_width)

mean_sepal_width = mean(sepal_width)
print "\nMean Sepal Width"
print mean_sepal_width

var_sepal_width = varian(sepal_width,mean_sepal_width)
print "Varian Sepal Width"
print var_sepal_width
 

# --- Petal Length
petal_length = [row[2] for row in dataIris] 
#print (petal_length)

mean_petal_length = mean(petal_length)
print "\nMean Petal Length"
print mean_petal_length

var_petal_length = varian(petal_length,mean_petal_length)
print "Varian Petal Length"
print var_petal_length


# --- Petal Width

petal_width = [row[3] for row in dataIris] 
#print (petal_width)
mean_petal_width = mean(petal_width)
print "\nMean Petal Width"
print mean_petal_width

var_petal_width = varian(petal_width,mean_petal_width)
print "Varian Petal Width"
print var_petal_width
```

```python
#3.

import matplotlib.pyplot as plt

plt.scatter(petal_length,petal_width)
plt.title('scatter plot 2D dari data petal length dan petal width')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
#4.

def data_center(listKolom,mean):
    dCenter = []
    for y in listKolom:
        operasiDat = y - mean
        dCenter.append(operasiDat)
    return dCenter

# -- List Data Center Petal Length
data_center_petal_length = data_center(petal_length, mean_petal_length)
#print data_central_petal_length

# -- List Data Center Petal Width
data_center_petal_width = data_center(petal_width, mean_petal_width)
#print data_central_petal_width

CenterDataMatrixPetal = []

for i in range(len(data_center_petal_length)):
    templist = [data_center_petal_length [i], data_center_petal_width [i]]
    CenterDataMatrixPetal.append(templist)
    
print CenterDataMatrixPetal
```

```python
#5.

plt.scatter(data_center_petal_length,data_center_petal_width)
plt.title('scatter plot 2D dari centered data matrix')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
#6.

def data_center(listKolom,mean):
    dCenter = []
    for y in listKolom:
        operasiDat = y - mean
        dCenter.append(operasiDat)
    return dCenter

mean_data_center_petal_length = sum(data_center_petal_length)/len(data_center_petal_length)
mean_data_center_petal_width = sum(data_center_petal_width)/len(data_center_petal_width)

print "\nMean Data Center Petal Length"
print mean_data_center_petal_length
print "Mean Data Center Petal Width"
print mean_data_center_petal_width


y = iris.target

plt.scatter(data_center_petal_length,data_center_petal_width, c=y, edgecolors='k')
    
plt.grid(True)
plt.title('scatter plot 2D dari centered data matrix')
plt.xlabel('Data Center Petal Length')
plt.ylabel('Data Center Petal Width')
plt.show()

print 'Legend'
print 'Warna ungu = Setosa'
print 'Warna hijau = Versicolor'
print 'Warna kuning = Virginica'
```

```python
#7.

def covarian(list1, mean1, list2, mean2):
    listcov = []
    for x in range(len(list1)):
        op1 = list1[x]-mean1        
        op2 = list2[x]-mean2
        opcov = op1*op2
        listcov.append(opcov)
        
    meanCov = sum(listcov)/len(listcov)
    
    return meanCov

var_datCen_petal_length = varian(data_center_petal_length,mean_data_center_petal_length)
var_datCen_petal_width = varian(data_center_petal_width, mean_data_center_petal_width)

covarian_datCen_Petal_width_length = covarian(data_center_petal_length,mean_data_center_petal_length,data_center_petal_width,mean_data_center_petal_width)

matrix_cov_datCen_petal = [[var_datCen_petal_length,covarian_datCen_Petal_width_length],[covarian_datCen_Petal_width_length, var_datCen_petal_width]]

print "\nCovariant matrix petal length dan petal width dari centered data matrix"
print matrix_cov_datCen_petal
```

11. Dengan satu kali percobaan, hitunglah probabilitas dimana Anda dapat memilih Iris yang memiliki sepal width lebih dari 3 cm!
12. Dengan 30 kali percobaan, hitunglah probabilitas dimana Anda dapat memilih 5 Iris yang memiliki sepal width lebih dari 3 cm!

Jawaban:

```python
#11.

listPro_sepal_width = []
for t in sepal_width:
    if t > 3:
        listPro_sepal_width.append(t)


probabilitas_sepal_width = float(len(listPro_sepal_width)) / float(len(sepal_width))

print "Probabilitas sepal width"
print probabilitas_sepal_width
```

```python
#12.
def faktorial(n):
    i = 1
    hasil = 1
    while i <= n:
        #kalikan hasil ke i simpan kembali ke hasil
        hasil = hasil * i
        # naikan nilai i sebesar 1
        i = i + 1
    return hasil
    
    

m = 30
k = 5
mk = m-k


kFak = faktorial(k)
mFak = faktorial(m)
mkFak = faktorial(mk)

combMK = mFak/(kFak*(mkFak))

#print combMK

probAnswer = (combMK)*(probabilitas_sepal_width**k) * ((1-probabilitas_sepal_width)**(m-k))

print "Probabilitas"
print probAnswer
```
